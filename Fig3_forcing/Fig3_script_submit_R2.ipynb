```julia
using GMT
using MAT
using Statistics
#cd("D:\\on_going_proposal\\running\\science\\on_going_paper\\bloom_2018")
```

```julia
f=matopen("Fig3_data_R2.mat")
cx=read(f,"cx")
cy=read(f,"cy")
chl=read(f,"chl")
xi=read(f,"xi")
yi=read(f,"yi")
mld=read(f,"mld")
T=read(f,"T")
S=read(f,"S")
pden=read(f,"pden")
wx=read(f,"wx")
wy=read(f,"wy")
wcurl2=read(f,"wcurl2")
curlclim67=read(f,"curlclim67")
close(f)

wx_vec=vec(wx)
wy_vec=vec(wy)

vec_data=vec(curlclim67)*10^5
I=findall(vec_data .> 1); vec_data[I].=1
I=findall(vec_data .< -1); vec_data[I].=-1
G=mat2ds([wx_vec wy_vec vec_data]);

vec_data=vec(wcurl2)*10^5
I=findall(vec_data .> 1); vec_data[I].=1
I=findall(vec_data .< -1); vec_data[I].=-1
GW=mat2ds([wx_vec wy_vec vec_data]);

xi_vec=vec(xi)
yi_vec=vec(yi)
vec_data=vec(mld)
Gmld=mat2ds([xi_vec yi_vec vec_data]);
pden_15=mean(pden[:,:,1:4],dims=3)
vec_data=vec(pden_15)
Ibig=findall(vec_data.>24.0)
vec_data[Ibig].=24
Gden=mat2ds([xi_vec yi_vec vec_data]);
temp_15=mean(T[:,:,1:4],dims=3)
vec_data=vec(temp_15)
Gtem=mat2ds([xi_vec yi_vec vec_data]);
salt_15=mean(S[:,:,1:4],dims=3)
vec_data=vec(salt_15)
Gsal=mat2ds([xi_vec yi_vec vec_data]);

cx_vec=vec(cx)
cy_vec=vec(cy)
vec_data=vec(chl)
Gchl=mat2ds([cx_vec cy_vec vec_data]);
```

```julia
x1=150; x2=210; y1=7; y2=30;
# gmtset(FONT_TAG="55p")
fx=12; ft=16;
subplot(grid="4x1", dims=(panels=(10,18.5)), margins="0.5c, 0.2c", 
        autolabel=(fill=:white,offset=0.5), 
        col_axes=(bott="",label=""),name="Fig3_R2_300dpi.tif",dpi=300)
    subplot(:set,panel=1)
    basemap(region=(x1,x2,y1,y2),proj=:Mercator, 
            yaxis=(annot=5, ticks=2.5), xaxis=(annot=10, ticks=5),
            conf=(MAP_FRAME_TYPE="plain",MAP_FRAME_PEN="thickest",
                  MAP_TICK_LENGTH_PRIMARY=0.2,FONT_ANNOT_PRIMARY=fx,FONT_TAG=ft))
    C=makecpt(cmap=:polar,range=(-1,1,0.2))
    contourf!(G,C)
    colorbar!(pos=(map=true, anchor=(213,7), size=(4.0,0.5)),
              frame=(annot=0.4,ticks=0.2),ylabel="x10@+-5@+ s@+-1@+",
              conf=(MAP_TICK_LENGTH_PRIMARY=0.2,FONT_ANNOT_PRIMARY=fx))
    contour!(Gchl, contours=[0.08],pen=("thicker",:black,:solid))
    coast!(shore=:thinnest,land=:grey)

    subplot(:set,panel=2)
    basemap(region=(x1,x2,y1,y2),proj=:Mercator, 
            yaxis=(annot=5, ticks=2.5), xaxis=(annot=10, ticks=5),
            conf=(MAP_FRAME_TYPE="plain",MAP_FRAME_PEN="thickest",
                  MAP_TICK_LENGTH_PRIMARY=0.2,FONT_ANNOT_PRIMARY=fx,FONT_TAG=ft))
    C=makecpt(cmap=:polar,range=(-1,1,0.2))
    contourf!(GW,C)
    colorbar!(pos=(map=true, anchor=(213,7), size=(4.0,0.5)),
              frame=(annot=0.4,ticks=0.2),ylabel="x10@+-5@+ s@+-1@+",
              conf=(MAP_TICK_LENGTH_PRIMARY=0.2,FONT_ANNOT_PRIMARY=fx))
    contour!(Gchl, contours=[0.08],pen=("thicker",:black,:solid))
    coast!(shore=:thinnest,land=:grey)

    subplot(:set,panel=3)
    basemap(region=(x1,x2,y1,y2),proj=:Mercator, 
            yaxis=(annot=5, ticks=2.5), xaxis=(annot=10, ticks=5),
            conf=(MAP_FRAME_TYPE="plain",MAP_FRAME_PEN="thicker",
                  MAP_TICK_LENGTH_PRIMARY=0.2,FONT_ANNOT_PRIMARY=fx,FONT_TAG=ft))
    C=makecpt(cmap=:jet,range=(0,80,10))
    contourf!(Gmld,C)
    colorbar!(pos=(map=true, anchor=(213,7), size=(4.0,0.5)),
              frame=(annot=20,ticks=10),ylabel="m",
              conf=(MAP_TICK_LENGTH_PRIMARY=0.2,FONT_ANNOT_PRIMARY=fx))
    contour!(Gchl, contours=[0.08],pen=("thicker",:black,:solid))
    coast!(shore=:thin,land=:grey)

    subplot(:set,panel=4)
    basemap(region=(x1,x2,y1,y2),proj=:Mercator, 
            yaxis=(annot=5, ticks=2.5), xaxis=(annot=10, ticks=5),
            conf=(MAP_FRAME_TYPE="plain",MAP_FRAME_PEN="thicker",
                  MAP_TICK_LENGTH_PRIMARY=0.2,FONT_ANNOT_PRIMARY=fx,FONT_TAG=ft))
    C=makecpt(cmap=:jet,range=(21,24,0.25))
    contourf!(Gden,C)
    colorbar!(pos=(map=true, anchor=(213,7), size=(4.0,0.5)),
              frame=(annot=1,ticks=0.25),ylabel="kg m@+-3@+",
              conf=(MAP_TICK_LENGTH_PRIMARY=0.2,FONT_ANNOT_PRIMARY=fx))
    contour!(Gchl, contours=[0.08],pen=("thicker",:black,:solid))
    coast!(shore=:thin,land=:grey)


subplot(:show)
```
