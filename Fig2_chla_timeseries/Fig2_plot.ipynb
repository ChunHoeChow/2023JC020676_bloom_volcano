```julia
# f=matopen("Fig2_data_smallerbox.mat")
f=matopen("Fig2_data.mat")
t=read(f,"x")
chl_mean=read(f,"chl_mean")
# chl_std=read(f,"chl_std")
area04=read(f,"area04")
aot_mean=read(f,"aot_mean")
rain_mean=read(f,"rain_mean")
gt=read(f,"gt")
# chl_mon_totalavg_time=read(f,"chl_mon_totalavg_time")
# chl_mon_totalstd_time=read(f,"chl_mon_totalstd_time")
close(f)

area04=area04/100000/1000/1000
L=length(t)
t_num=zeros(size(t))
for m=1:L    
    ## convert date to datenumber in seconds needed by GMT plot to show the xlabel in dateformat
    t_num[m]=(Dates.value(Date(gt[m,1],gt[m,2],gt[m,3]))-Dates.value(Date(1970,1,1)))*86400
end

f=matopen("chl_smaller_box_timemean_std_max_2003_2018.mat")
chl_timemean=read(f,"chl_timemean")
# chl_timestd=read(f,"chl_timestd")
# chl_timemax=read(f,"chl_timemax")
year=read(f,"year")
month=read(f,"month")
chl_box2_spacemean=read(f,"chl_box2_spacemean")
chl_box2_spacestd=read(f,"chl_box2_spacestd")
close(f)
L=length(year)
t_num2=zeros(size(year))
for m=1:L    
    ## convert date to datenumber in seconds needed by GMT plot to show the xlabel in dateformat
    t_num2[m]=(Dates.value(Date(year[m],month[m],15))-Dates.value(Date(1970,1,1)))*86400
end

f=matopen("whiskers_data_170_190_18_27_2018.mat")
# x=read(f,"time2")
q1=read(f,"Q1")
q2=read(f,"Q2")
q3=read(f,"Q3")
wup=read(f,"wup")
wdn=read(f,"wdn")
close(f)

f=matopen("whiskers_data_170_190_18_27_2003_2018.mat")
# x=read(f,"time2")
q1_all=read(f,"Q1")
q2_all=read(f,"Q2")
q3_all=read(f,"Q3")
wup_all=read(f,"wup")
wdn_all=read(f,"wdn")
close(f)

f=matopen("whiskers_mondata_170_190_18_27_2003_2018.mat")
# x=read(f,"time2")
q1_mon=read(f,"Q1")
q2_mon=read(f,"Q2")
q3_mon=read(f,"Q3")
wup_mon=read(f,"wup")
wdn_mon=read(f,"wdn")
# data_mon=read(f,"all_data")
close(f)
```
